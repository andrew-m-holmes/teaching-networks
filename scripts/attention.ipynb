```python
from transformers import AutoTokenizer, AutoModel
from bertviz import head_view, neuron_view
```

```python
tokenizer = AutoTokenizer.from_pretrained("openai-community/gpt2")
model = AutoModel.from_pretrained("openai-community/gpt2")
```

```python
sentence = "What do you prefer, pancakes or waffles for breakfast?"
inputs = tokenizer.encode(sentence, return_tensors="pt", add_special_tokens=True)
outputs = model(inputs, output_attentions=True)
attention = outputs[-1]
tokens = [tokenizer.decode(inid, skip_special_tokens=False) for inid in inputs[0]]
```

```python
head_view(attention, tokens)
```
